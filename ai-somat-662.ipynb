```python
import math

# 1. 설계 조건 입력
torque_Nm = 1500           # 외부에서 작용하는 비틀림 모멘트 (N·m)
yield_strength = 250       # 재료의 항복강도 (MPa)
target_safety_factor = 2.5 # 목표 안전율 (SF)

# 2. 허용 응력 계산 (항복강도를 안전율로 나눔)
# 전단 항복강도는 인장 항복강도의 약 0.577배 (폰 미세스 기준 단순화)
allowable_shear_stress = (yield_strength * 0.577) / target_safety_factor

# 3. 최적 형상(지름) 역산 공식 적용
# 전단응력(Tau) = (16 * T) / (pi * d^3) 공식을 지름(d)에 대해 정리
# 단위 맞춤: N·m -> N·mm ( * 1000 )
torque_Nmm = torque_Nm * 1000
optimal_d_cubed = (16 * torque_Nmm) / (math.pi * allowable_shear_stress)
optimal_d = optimal_d_cubed ** (1/3)

# 4. 결과 출력
print("--- 최적 설계 치수 역산 결과 ---")
print(f"목표 안전율: {target_safety_factor}")
print(f"허용 전단 응력: {allowable_shear_stress:.2f} MPa")
print(f"필요한 최소 축 지름: {optimal_d:.2f} mm")

# 실무 적용 팁: 구한 최소 지름보다 큰 KS 규격 치수로 최종 선정합니다.
standard_d = math.ceil(optimal_d / 5) * 5  # 5mm 단위로 올림 처리
print(f"최종 추천 규격 지름: {standard_d} mm")
```

```python
import math

# 1. 변수 선언 및 입력 데이터
load_kN = 50.0               # 작용하는 인장 하중 (kN)
yield_strength_MPa = 250.0   # 재료의 항복 강도 (MPa)

# 단위 변환 (N, Pa 단위로 통일)
load_N = load_kN * 1000.0
yield_strength_Pa = yield_strength_MPa * 1e6

# 평가할 안전율 리스트
safety_factors = [1.5, 2.0, 2.5, 3.0]

print("=== 안전율에 따른 최소 지름 최적 설계 결과 ===")

# 2. 안전율별 반복 계산
for sf in safety_factors:
    # 허용 응력 계산
    allowable_stress_Pa = yield_strength_Pa / sf

    # 최소 단면적 계산 (A = P / sigma_allow)
    min_area_m2 = load_N / allowable_stress_Pa

    # 최소 지름 계산 (d = sqrt(4 * A / pi))
    min_diameter_m = math.sqrt((4 * min_area_m2) / math.pi)

    # 결과를 mm 단위로 변환
    min_diameter_mm = min_diameter_m * 1000.0

    # 3. 결과 출력
    print(f"안전율 {sf:.1f} 적용 시 -> 허용 응력: {allowable_stress_Pa/1e6:.1f} MPa, 요구되는 최소 지름: {min_diameter_mm:.2f} mm")
```

```python
import math

# 1. 하중 및 재료 물성치 설정
load_N = 50000          # 축 방향 인장 하중 (N)
yield_strength = 250    # 재료 항복강도 (MPa)
safety_factor = 2.0     # 목표 안전율

# 2. 허용 응력 계산
allowable_stress = yield_strength / safety_factor

# 3. 최소 중량 설계 (지름 최적화) 알고리즘 적용
diameter_mm = 1.0       # 가상의 초기 지름 1mm부터 탐색 시작

while True:
    area = (math.pi * diameter_mm**2) / 4
    current_stress = load_N / area

    # 발생 응력이 허용 응력보다 작거나 같아지면 탐색 종료
    if current_stress <= allowable_stress:
        break

    diameter_mm += 0.5  # 0.5mm씩 지름을 점진적으로 늘려가며 반복 확인

# 4. 결과 출력
print(f"목표 안전율({safety_factor}) 만족을 위한 최소 설계 지름: {diameter_mm:.1f} mm")
print(f"이때 부품에 발생하는 실제 응력: {current_stress:.1f} MPa")
```
