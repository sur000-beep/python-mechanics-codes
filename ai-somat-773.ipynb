```python
import math

# 1. 변수 선언 및 입력 데이터 (SI 단위계 통일)
diameter_mm = 30.0         # 축의 지름 (mm)
length_m = 1.0             # 축의 길이 (m)
torque_Nm = 500.0          # 작용하는 토크 (N*m)
G_GPa = 80.0               # 전단탄성계수 (GPa)

# 2. 단위 변환 (m, N, Pa 단위로 통일)
d_m = diameter_mm / 1000.0 # 지름 (m)
r_m = d_m / 2.0            # 반지름 (m)
G_Pa = G_GPa * 1e9         # 전단탄성계수 (Pa)

# 3. 극관성모멘트(J) 계산
J = (math.pi * d_m**4) / 32.0

# 4. 최대 전단 응력(Tau_max) 계산
tau_max_Pa = (torque_Nm * r_m) / J
tau_max_MPa = tau_max_Pa / 1e6  # MPa 단위로 변환

# 5. 비틀림 각(Theta) 계산
# 공식: Theta = (T * L) / (G * J) -> 결과는 라디안(Radian)
theta_rad = (torque_Nm * length_m) / (G_Pa * J)

# 라디안을 우리가 익숙한 도(Degree) 단위로 변환
theta_deg = math.degrees(theta_rad)

# 6. 결과 출력
print("=== 원형 축의 비틀림 해석 결과 ===")
print(f"극관성모멘트(J): {J:.8e} m^4")
print(f"발생하는 최대 전단 응력: {tau_max_MPa:.2f} MPa")
print(f"축의 비틀림 각: {theta_deg:.2f} 도 (Degree)")
```
