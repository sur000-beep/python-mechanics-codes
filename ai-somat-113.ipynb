```python
import math

# 1. 변수 선언 및 주어진 데이터 입력
diameter_mm = 50.0  # 축의 지름 (mm)
load_kN = 10.0      # 인장 하중 (kN)

# 2. 단위 변환 (SI 단위계: m, N으로 통일)
diameter_m = diameter_mm / 1000.0  # mm를 m로 변환
load_N = load_kN * 1000.0          # kN을 N으로 변환

# 3. 단면적 및 수직 응력 계산
# 원의 단면적 계산 (A = pi * d^2 / 4)
area_m2 = (math.pi * diameter_m**2) / 4.0

# 응력 계산 (Pa 단위)
stress_Pa = load_N / area_m2

# 결과를 보기 쉽게 MPa 단위로 변환 (1 MPa = 10^6 Pa)
stress_MPa = stress_Pa / 1e6

# 4. 결과 출력
print("=== 재료역학 응력 계산 결과 ===")
print(f"강철 축의 단면적: {area_m2:.6f} m^2")
print(f"발생하는 수직 응력: {stress_MPa:.2f} MPa")
```
