```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 변수 선언 및 입력 데이터
L = 4.0      # 보의 전체 길이 (m)
P = 20.0     # 중앙에 작용하는 집중 하중 (kN)

# 보의 길이 방향으로 400개의 지점 생성 (0부터 L까지)
x = np.linspace(0, L, 400)
V = np.zeros_like(x)
M = np.zeros_like(x)

# 2. 구간별 전단력(V) 및 굽힘 모멘트(M) 자동 계산
for i, xi in enumerate(x):
    if xi < L/2:
        V[i] = P / 2
        M[i] = (P / 2) * xi
    else:
        V[i] = -P / 2
        M[i] = (P / 2) * (L - xi)

# 대표 결과값 화면 출력
print("=== 보의 해석 결과 대표값 ===")
print(f"최대 전단력 (Max Shear Force): {P/2:.1f} kN")
print(f"최대 굽힘 모멘트 (Max Bending Moment): {(P*L)/4:.1f} kN·m (위치: x = {L/2:.1f} m)")

# 3. SFD 및 BMD 그래프 시각화
plt.figure(figsize=(10, 6))

# 첫 번째 그래프: SFD
plt.subplot(2, 1, 1)
plt.plot(x, V, 'r-', linewidth=2)
plt.fill_between(x, V, color='red', alpha=0.1)
plt.title('Shear Force Diagram (SFD)')
plt.ylabel('Shear Force (kN)')
plt.grid(True)

# 두 번째 그래프: BMD
plt.subplot(2, 1, 2)
plt.plot(x, M, 'b-', linewidth=2)
plt.fill_between(x, M, color='blue', alpha=0.1)
plt.title('Bending Moment Diagram (BMD)')
plt.xlabel('Beam Length (m)')
plt.ylabel('Bending Moment (kN·m)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
