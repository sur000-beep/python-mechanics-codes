```python
import sympy as sp

# 1. 미지수 기호(Symbol) 정의
# Ax: A점 수평 반력, Ay: A점 수직 반력, By: B점 수직 반력
Ax, Ay, By = sp.symbols('Ax Ay By')

# 2. 주어진 조건 입력
L = 5.0      # 보의 전체 길이 (m)
P = 10.0     # 집중 하중 (kN)
a = 2.0      # 하중이 작용하는 위치 (A점으로부터의 거리, m)

# 3. 평형 방정식 설정 (방정식 = 0 형태로 작성)
# 수평 방향 힘의 합 = 0
eq_fx = Ax

# 수직 방향 힘의 합 = 0 (위로 향하는 힘은 +, 아래는 -)
eq_fy = Ay + By - P

# A점 기준 모멘트의 합 = 0 (반시계 방향 +, 시계 방향 -)
# 하중 P는 시계 방향 모멘트, 반력 By는 반시계 방향 모멘트 발생
eq_moment_A = By * L - P * a

# 4. 연립방정식 풀이
# sp.solve([방정식 리스트], (미지수 튜플))
solutions = sp.solve([eq_fx, eq_fy, eq_moment_A], (Ax, Ay, By))

# 5. 결과 출력
print("=== 정역학 평형 방정식 해석 결과 ===")
print(f"A점 수평 반력 (Ax): {solutions[Ax]:.1f} kN")
print(f"A점 수직 반력 (Ay): {solutions[Ay]:.1f} kN")
print(f"B점 수직 반력 (By): {solutions[By]:.1f} kN")
```

```python
import numpy as np

# 1. 평형 방정식의 계수 행렬 [A] 생성 (X방향, Y방향 힘의 합력 계수)
# 예시: 3개의 미지 부재력(F1, F2, F3)을 가진 단순 트러스 격점
# cos(45), sin(45) 등의 기하학적 각도 데이터를 입력합니다.
cos45 = np.cos(np.radians(45))
sin45 = np.sin(np.radians(45))

A = np.array([
    [1.0, cos45, 0.0],   # 격점 A의 X방향 평형
    [0.0, sin45, 1.0],   # 격점 A의 Y방향 평형
    [0.0, 0.0,  -1.0]    # 격점 B의 Y방향 평형 (예시)
])

# 2. 하중 벡터 {B} 생성 (외부에서 가해지는 힘, kN 단위)
B = np.array([0, 10, -5]) # 우변의 상수항 (외력)

# 3. 행렬 연산을 통한 미지 부재력 {x} 즉시 계산
forces = np.linalg.solve(A, B)

# 4. 결과 출력
print("--- 트러스 부재력 계산 결과 ---")
print(f"부재력 F1: {forces[0]:.2f} kN")
print(f"부재력 F2: {forces[1]:.2f} kN")
print(f"부재력 F3: {forces[2]:.2f} kN")
```
