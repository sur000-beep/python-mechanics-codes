```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 입력 데이터 (평면 응력 상태, 단위: MPa)
sigma_x = 80.0
sigma_y = 40.0
tau_xy = 30.0

# 2. 모어의 원 중심(C)과 반지름(R) 계산
C = (sigma_x + sigma_y) / 2.0
R = np.sqrt(((sigma_x - sigma_y) / 2.0)**2 + tau_xy**2)

# 3. 주응력 및 최대 전단 응력 계산
sigma_1 = C + R
sigma_2 = C - R
tau_max = R

# 수치 결과 출력
print("=== 모어의 원(Mohr's Circle) 해석 결과 ===")
print(f"원의 중심 (C): {C:.1f} MPa")
print(f"원의 반지름 (R): {R:.1f} MPa")
print(f"최대 주응력 (sigma_1): {sigma_1:.1f} MPa")
print(f"최소 주응력 (sigma_2): {sigma_2:.1f} MPa")
print(f"최대 전단 응력 (tau_max): {tau_max:.1f} MPa")

# 4. 시각화 (그래프 그리기)
theta = np.linspace(0, 2 * np.pi, 100)
circle_x = C + R * np.cos(theta)
circle_y = R * np.sin(theta)

plt.figure(figsize=(8, 8))
# 원 그리기
plt.plot(circle_x, circle_y, 'b-', label="Mohr's Circle")
# 가로축(주응력 선) 표시
plt.plot([sigma_2, sigma_1], [0, 0], 'ko-', label='Principal Stresses')
# 초기 응력 상태(X-Y면) 표시
plt.plot([sigma_x, sigma_y], [-tau_xy, tau_xy], 'ro-', label='Initial State (X-Y)')

# 축 설정 및 꾸미기
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlabel('Normal Stress (MPa)')
plt.ylabel('Shear Stress (MPa)')
plt.title("Mohr's Circle for Plane Stress")
plt.grid(True)
plt.legend()
plt.axis('equal') # 원이 찌그러지지 않도록 가로세로 비율 동일하게 설정
plt.show()
```
