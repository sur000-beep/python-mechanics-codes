```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials

# 1. 구글 API 인증 설정 및 시트 열기
scope = ["https://spreadsheets.google.com/feeds", "https://www.googleapis.com/auth/drive"]
# 발급받은 JSON 키 파일 이름을 입력합니다
creds = ServiceAccountCredentials.from_json_keyfile_name("/content/my-project-5-9-2026-uhsur-6aa6852b8e40.json", scope)
client = gspread.authorize(creds)

# 생성한 구글 스프레드시트의 정확한 제목을 입력하여 문서 열기
doc = client.open("재료역학_해석_보고서")
sheet = doc.get_worksheet(0) # 첫 번째 탭(워크시트) 선택

# 2. 본문에서 계산된 파이썬 결과 데이터 변수 예시
part_name = "구동축 (Shaft_01)"
stress_result = 145.2  # 단위: MPa
safety_factor = 2.1
status = "안전 (Pass)"
# 3. 구글 스프레드시트에 데이터 추가 (새로운 행으로 누적 적재)
# 기존 데이터 아래에 순서대로 데이터가 저장됩니다.
sheet.append_row([part_name, stress_result, safety_factor, status])
print("🚀 구글 스프레드시트로 실시간 보고서 데이터 전송 성공!")
```
