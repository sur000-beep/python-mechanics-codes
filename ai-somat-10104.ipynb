```python
# 1. AI 또는 3D FEA 해석 프로그램이 도출한 결과값
fea_stress_MPa = 450.0

# 2. 직관을 활용한 단순화 모델(외팔보) 파라미터
force_N = 1000.0       # 끝단 하중 (N)
length_m = 0.1         # 부품 길이 (100mm = 0.1m)
width_m = 0.02         # 부품 폭 (20mm = 0.02m)
height_m = 0.01        # 부품 두께 (10mm = 0.01m)

# 3. 단순 굽힘 공식에 의한 손계산(Hand Calculation) 수행
# 굽힘 모멘트 (M = P * L)
moment_Nm = force_N * length_m

# 직사각형 단면계수 (Z = b * h^2 / 6)
section_modulus_m3 = (width_m * height_m**2) / 6.0

# 손계산 굽힘 응력 (Sigma = M / Z)
hand_calc_stress_Pa = moment_Nm / section_modulus_m3
hand_calc_stress_MPa = hand_calc_stress_Pa / 1e6

# 4. 오차율 비교 및 Sanity Check 로직
error_margin_pct = abs(fea_stress_MPa - hand_calc_stress_MPa) / hand_calc_stress_MPa * 100.0

print("=== AI 결과 검증(Sanity Check) 시스템 ===")
print(f"FEA/AI 해석 결과: {fea_stress_MPa:.1f} MPa")
print(f"단순화 모델 손계산 결과: {hand_calc_stress_MPa:.1f} MPa")
print(f"오차율: {error_margin_pct:.1f}%")

print("-" * 40)
# 오차율이 20%를 넘어가면 모델링(입력)에 치명적 오류가 있을 가능성이 높음
if error_margin_pct <= 20.0:
    print("검증 완료: AI 해석 결과가 물리적 직관(손계산) 범위 내에 있습니다. (정상)")
else:
    print("경고(WARNING): AI 결과와 손계산의 차이가 너무 큽니다!")
    print("-> 하중 방향, 단위(Unit) 설정, 또는 경계 조건을 다시 확인하세요 (GIGO 주의).")
```
