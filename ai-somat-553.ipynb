```python
import numpy as np

# 1. 주어진 데이터 및 요소 물성치 입력
k1 = 50.0  # 첫 번째 스프링(요소 1)의 강성 (N/mm)
k2 = 50.0  # 두 번째 스프링(요소 2)의 강성 (N/mm)
F3 = 100.0 # 절점 3에 작용하는 외력 (N)

# 2. 전체 강성 행렬(Global Stiffness Matrix) 초기화 (절점이 3개이므로 3x3 행렬)
K_global = np.zeros((3, 3))

# 3. 각 요소의 강성 행렬을 전체 행렬에 조립 (Assembly)
# 요소 1 (절점 1, 2 연결)
K_global[0:2, 0:2] += np.array([[k1, -k1],
                                [-k1, k1]])

# 요소 2 (절점 2, 3 연결)
K_global[1:3, 1:3] += np.array([[k2, -k2],
                                [-k2, k2]])

# 4. 경계 조건(Boundary Condition) 적용
# 절점 1은 벽에 고정되어 변위가 0 (u1 = 0)
# 변위가 0인 절점의 행과 열을 제거하여 수정된 행렬(Reduced Matrix) 생성
K_reduced = K_global[1:3, 1:3]
F_reduced = np.array([0, F3]) # 절점 2에는 외력이 없고, 절점 3에는 F3 작용

# 5. 연립방정식 풀이 ( {U} = [K]^-1 * {F} )
# 절점 2와 절점 3의 변위 계산
U_reduced = np.linalg.solve(K_reduced, F_reduced)

# 전체 변위 벡터 구성 (u1 = 0 추가)
U_global = np.array([0, U_reduced[0], U_reduced[1]])

# 6. 결과 출력
print("=== 1D 유한요소해석(FEA) 결과 ===")
print(f"절점 1의 변위: {U_global[0]:.2f} mm (고정단)")
print(f"절점 2의 변위: {U_global[1]:.2f} mm")
print(f"절점 3의 변위: {U_global[2]:.2f} mm")
```

```python
import numpy as np

# 1. 요소 강성 정의 (스프링 상수 k = AE/L)
k1 = 1000  # 첫 번째 요소의 강성 (N/m)
k2 = 2000  # 두 번째 요소의 강성 (N/m)

# 2. 전체 강성 행렬(Global Stiffness Matrix) 조립
# 3개의 절점을 가진 2개의 요소가 결합된 3x3 행렬
K = np.array([
    [k1, -k1, 0],
    [-k1, k1 + k2, -k2],
    [0, -k2, k2]
])

# 3. 하중 벡터 설정 및 경계 조건(고정점) 적용
F = np.array([0, 0, 500])  # 마지막 3번 절점에 500N의 힘이 작용

# 1번 절점은 벽에 단단히 고정되어 변위가 0이므로, 행렬 계산에서 1번 행과 열을 제외(축소)합니다.
K_reduced = K[1:, 1:]
F_reduced = F[1:]

# 4. 변위 계산 (선형 대수 방정식 풀이: [K]{u} = {F})
u_reduced = np.linalg.solve(K_reduced, F_reduced)

# 고정점의 변위(0)를 다시 포함하여 전체 변위 결과 출력
u = np.concatenate(([0], u_reduced))

print("--- 유한요소해석(FEA) 변위 계산 결과 ---")
print(f"절점 1 (벽 고정) 이동량 : {u[0]:.4f} m")
print(f"절점 2 (중간 점) 이동량  : {u[1]:.4f} m")
print(f"절점 3 (하중 작용점) 이동량: {u[2]:.4f} m")
```
